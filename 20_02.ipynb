```python
import matplotlib.pyplot as plt
```

```python
car_make_count=car['Car Model'].value_counts()
#craete a pie chart
#fig as figure
#ax means attribute
fig,ax=plt.subplots(figsize=(4,4))
ax.pie(car_make_count,labels=car_make_count.index,autopct='%1.0f%%',startangle=90)
ax.axis('equal')
ax.set_title("Distribution of car makes")
plt.show()
```

```python
caryear_counts=car['Car Year'].value_counts()
fig,ax=plt.subplots(figsize=(5,8))
ax.pie(caryear_counts,labels=caryear_counts.index,autopct='%1.0f%%',startangle=90)
ax.axis('equal')
plt.show()
```

```python
#is there relationship b/w  the car make and the sale price?
plt.scatter(car['Car Make'],car['Sale Price'])
plt.xlabel('Car Make')
plt.ylabel('Sale Price')
plt.show()
```

```python
# prompt:

# Create a DataFrame with the student data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Java': [95, 80, 75, 90, 85],
    'Python': [85, 90, 70, 80, 95],
    'Flutter': [70, 80, 90, 75, 85]
}
df = pd.DataFrame(data)

# Calculate the total marks for each student
df['Total Marks'] = df['Java'] + df['Python'] + df['Flutter']

# Find the student with the highest total marks
student_with_highest_marks = df[df['Total Marks'] == df['Total Marks'].max()]

# Print the student with the highest total marks
print(student_with_highest_marks)
```

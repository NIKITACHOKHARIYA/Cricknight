```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfile=pd.read_csv('/content/housing.csv')
dfile.head()
```

```python
X=dfile.drop('Price',axis=1)
X.head()
```

```python
Y=dfile['Price']
Y
```

```python
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train.head()
```

```python
lr=LinearRegression() #create our linear model

#fitting the model on the training data and try to predict the X_test
```

```python
h_model=lr.fit(X_train,Y_train)
predict=h_model.predict(X_test)
```

```python
print(predict)
h_model
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,predict)
```
